```python
import anndata as ad
import pandas as pd
import scanpy as sc
import seaborn as sns
import matplotlib.pyplot as plt
#import colorcet as cc
```

```python
#RCC
adataRCC = ad.read_h5ad("/home/larissa/Documents/Masterarbeit/objects/kidney_cancer_processed.h5ad")


sc.pp.normalize_per_cell(adataRCC,counts_per_cell_after=1e4)
sc.pp.log1p(adataRCC)
```

```python
#MF
adataMF = ad.read_h5ad("/home/larissa/Documents/Masterarbeit/objects/MF_subcluster_subset.h5ad")
adataMF_sub = adataMF[(adataMF.obs["treatment"] == "Control")  |  (adataMF.obs["treatment"] == "MF_pre_treatment")]
```

```python
#cell type counts
#RCC

#Tumor
for clust in list(adataRCC.obs["cell_type"].unique()):
    adataRCC_sub = adataRCC[adataRCC.obs["accLabel"] == "T"]
    adataRCC_sub = adataRCC_sub[adataRCC_sub.obs["cell_type"] == clust]
    print(clust, adataRCC_sub.obs.count())

#Necrosis
for clust in list(adataRCC.obs["cell_type"].unique()):
    adataRCC_sub = adataRCC[adataRCC.obs["accLabel"] == "T"]
    adataRCC_sub = adataRCC_sub[adataRCC_sub.obs["necrosis"] == "Positive"]
    adataRCC_sub = adataRCC_sub[adataRCC_sub.obs["cell_type"] == clust]
    print(clust, adataRCC_sub.obs.count())

#Healthy
for clust in list(adataRCC.obs["cell_type"].unique()):
    adataRCC_sub = adataRCC[adataRCC.obs["accLabel"] == "N"]
    adataRCC_sub = adataRCC_sub[adataRCC_sub.obs["necrosis"] == "Positive"]
    adataRCC_sub = adataRCC_sub[adataRCC_sub.obs["cell_type"] == clust]
    print(clust, adataRCC_sub.obs.count())

#MF

#Tumor
for clust in list(adataMF_sub.obs["subcluster"].unique()):
    adataMF_subb = adataMF_sub[adataMF_sub.obs["treatment"] == "MF_pre_treatment"]
    adataMF_subb = adataMF_subb[adataMF_subb.obs["subcluster"] == clust]
    print(clust, adataMF_subb.obs.count())

#Healthy    
for clust in list(adataMF_sub.obs["subcluster"].unique()):
    adataMF_subb = adataMF_sub[adataMF_sub.obs["treatment"] == "Control"]
    adataMF_subb = adataMF_subb[adataMF_subb.obs["subcluster"] == clust]
    print(clust, adataMF_subb.obs.count())
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
res = sc.pl.violin(adataRCC, ["pct_counts_mito"], groupby="sample", ax = ax, show = False)
plt.axhline(20, color='red', linestyle='--', label='20% threshold')
plt.xlabel("Samples")
plt.ylabel("Mitochondrial % per cell")
#plt.title("Mitochondrial content after QC")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
#ax.set_ylabel("")        
plt.yticks([20], ["20%"])
ax.tick_params(left=False) 
plt.show()
plt.close()
```
